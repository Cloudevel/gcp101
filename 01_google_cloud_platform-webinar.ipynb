# *Google Cloud Platform*.



## Cmputo en la nube.

Seun el *NIST*, cmputo en la nube se define como un modelo para habilitar mediante un red ubicua, conveniente y bajo demanda el acceso a un conjunto de recursos de cmputo compartidos y configurables (redes, servidores, almacenamiento, aplicaciones y servicios) que pueden ser aprovisionados rpidamente y desplegados con un mnimo esfuerzo de gestin o de interaccin por parte del proveedor del servicio.

https://www.nist.gov/news-events/news/2011/10/final-version-nist-cloud-computing-definition-published

### Caractersticas del cmputo en la nube.

* Autoservicio bajo demanda.
* Acceso amplio a redes de comunicaciones.
* Acopio (pool) de recursos de cmputo.
* Rpida expansin o elasticidad de los recursos.
* Servicio medido.

### Modelos de servicio.

* Software como servicio (SaaS).
* Plataforma como servicio (PaaS).
* Infraestructura como servicio (IaaS).

### Modelos de despliegue.

* Nube privada.
* Nube comunitaria.
* Nube pblica.
* Nube hbrida.

## Tipos de servicios en la nube.

* Recursos de cmputo.
* Almacenamiento.
* Red.
* Servicios especializados.

## *Google Cloud Platfom*.

[*Google Cloud Platform* (*GCP*)](https://cloud.google.com/) es un conjunto de servicios globales de *PaaS* e *IaaS* sustentado por la red de comunicaciones y los centros de datos de *Google*, los cuales pueden funcionar bajo un esquema de nube pblica e hbrida.

Por otra parte *Google* tambin ofrece un conjunto de servicios *SaaS* llamado [*Google Workspace*](https://workspace.google.com/).

### Regiones y zonas.

La infaestructura  de *GCP* se encuentra distribuida por todo el mundo mediante una red de centros de datos delimitados por regiones y zonas. 

La siguiente liga muestra la informacin ms actualizada de las regiones y zonas que *GCP* ofrece, as como los servicios que soporta cada una de ellas.


https://cloud.google.com/about/locations

### Oferta de precios de *GCP*.

* Facturacin por segundo de recursos de cmputo y procesamiento de datos entre otros.
* Descuentos por uso continuo.
* Descuentos por uso comprometido.
* Aprovisionamiento de instancias de mquinas virtuales a la medida.
